# Two options for subplots

```python
# Subplots option 1
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2,
                                          ncols = 2,
                                          figsize = (10,5))

ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(x[:10], x[:10]);
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2 - list indexing
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                      figsize=(10,5))
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(x[:10], x[:10]);
ax[1,1].hist(np.random.randn(1000));
```

```python
# More advanced use the Object Oriented Method
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50 ]
over_50
```

```python
#Pyplot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target');
```

```python
# OO with pyplot Method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
# ax.set_xlim([45,100])
```

```python
over_50.head()
```

```python
# OO Method from Scratch
fig, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x = over_50["age"],
                    y = over_50["chol"],
                    c = over_50["target"])

ax.set(title=" Heart Disease Cholesterol Rate",
      xlabel="Age",
      ylabel="Cholesterol");
#unpack elements
ax.legend(*scatter.legend_elements(), title="Target");
ax.axhline(over_50["chol"].mean(), linestyle='--');
```

```python
# subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True);
# add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);
# Customize
ax0.set(title="Heart Disease Cholesterol Rate",
       ylabel="Cholesterol");

# add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target");

# Add a meanline
ax0.axhline(y=over_50["chol"].mean(), linestyle="--");


# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])
# Customize
ax1.set(title="Heart Disease Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate")

# add a legend to ax0
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--")
# Add a title to figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
